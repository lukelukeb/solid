```python
import os
def fasta_seq():
    file = input("Enter File name with .fasta for saving:")
    Entrez_id = input("Enter id:")
    filename = file
    if not os.path.isfile(filename):
        # Downloading...
        net_handle = Entrez.efetch(
            db="nucleotide", id=Entrez_id, rettype="fasta", retmode="text"
        )
        out_handle = open(filename, "w")
        out_handle.write(net_handle.read())
        out_handle.close()
        net_handle.close()
        print("Saved")

    print("Parsing...")
    record = SeqIO.read(filename, "fasta")
    print(record)
```

```python
fasta_seq()
```

```python
from Bio.Blast import NCBIWWW
blast = SeqIO.read("985567450.fasta", format="fasta")
result_handle = NCBIWWW.qblast("blastn", "nt", blast.seq) # blast is perform in nucleotide ddtabase and perse sequence
```

```python
with open("my_blast.xml", "w") as out_handle:
    out_handle.write(result_handle.read())
result_handle.close()
```

```python
from Bio.Blast import NCBIXML # NCBIXML read the xml record store in xml file that download as alignment file
result_handle = open("my_blast.xml")
blast_record = NCBIXML.read(result_handle)
E_VALUE_THRESH = 0.04
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            print("sequence:", alignment.title)
            print(hsp.query[0:75] + "...")
            print(hsp.match[0:75] + "...")
            print(hsp.sbjct[0:75] + "...")
```
